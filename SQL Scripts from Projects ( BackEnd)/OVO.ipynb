# [dbo].[SP_AddAutoOrderDtl]

```sql
/****** Object:  StoredProcedure [dbo].[SP_AddAutoOrderDtl]    Script Date: 27-10-2023 15:01:16 ******/
SET ANSI_NULLS ON
GO
SET QUOTED_IDENTIFIER ON
GO
CREATE PROCEDURE [dbo].[SP_AddAutoOrderDtl] (
	@Id INT = NULL
	,@OrderNo varchar(50)
	,@ProductId int
	,@Qty int
	,@Price float
	,@Discount float
	,@ItemValue float
	,@IsActive BIT = 1
	,@CreatedDate DATETIME
	,@CreatedByUserId INT
	,@UpdatedDate DATETIME
	,@UpdatedByUserId INT
	,@DiscountedAmt float
	,@ErrorNumber INT OUTPUT
	,@ErrorMessage NVARCHAR(4000) OUTPUT
	)
AS
BEGIN
	BEGIN TRANSACTION

	BEGIN TRY
		IF ISNULL(@Id,0)=0
		begin
			INSERT INTO [dbo].[OrderDtl] (
				 OrderNo
			    ,[ProductId]
			    ,[Qty]
			    ,[Price]
			    ,[Discount]
			    ,[ItemValue]
				,[IsActive]
				,[CreatedDate]
				,[CreatedByUserId]
				,[UpdatedDate]
				,[UpdatedByUserId]
				,[DiscountedAmt]
				)
			VALUES (
				 @OrderNo
				,@ProductId
				,@Qty
				,@Price
				,@Discount
				,@ItemValue
				,ISNULL(@IsActive, 1)
				,ISNULL(@CreatedDate, GETDATE())
				,@CreatedByUserId
				,ISNULL(@UpdatedDate, GETDATE())
				,@UpdatedByUserId
				,@DiscountedAmt
				)
			    --update Stock set AvailableStocks=AvailableStocks-@Qty where ProductId=@ProductId
				 declare @WarehouseId int
			   set @WarehouseId = (select WareHouseId from OrderHdr where OrderNo =@OrderNo)
			   update WareHouseStock set AvailableStock=AvailableStock-@Qty where ProductId=@ProductId and WareHouseId = @WarehouseId
				 --Removed in Cart Items after Ordered Placed
				 delete from Cart where ProductId=@ProductId and CreatedByUserId=@CreatedByUserId
				end
		ELSE
			UPDATE [dbo].[OrderDtl]
			SET  OrderNo = @OrderNo
			    ,[ProductId] = @ProductId
			    ,[Qty] = @Qty
			    ,[Price] = @Price
			    ,[Discount] = @Discount
			    ,[ItemValue] = @ItemValue
				,[IsActive] = @IsActive
				,[UpdatedDate] = ISNULL(@UpdatedDate, GETDATE()) AT TIME ZONE 'UTC' AT TIME ZONE 'India Standard Time'
				,[UpdatedByUserId] = @UpdatedByUserId
				,[DiscountedAmt] =@DiscountedAmt
			WHERE Id = @Id
	END TRY

	BEGIN CATCH
		SELECT @ErrorNumber = ERROR_NUMBER()
			,@ErrorMessage = ERROR_MESSAGE()

		IF @@TRANCOUNT > 0
			ROLLBACK TRANSACTION;
		RAISERROR ('Error: %d: %s', 16, 1, @ErrorNumber, @ErrorMessage) ;
	END CATCH;

	IF @@TRANCOUNT > 0
		COMMIT TRANSACTION;

	SET @ErrorNumber = 0
	SET @ErrorMessage = 'Success'
END
GO
```

```sql
/****** Object:  StoredProcedure [dbo].[SP_AddWalletTransaction]    Script Date: 27-10-2023 15:01:16 ******/
SET ANSI_NULLS ON
GO
SET QUOTED_IDENTIFIER ON
GO
CREATE Procedure [dbo].[SP_AddWalletTransaction]
(@Id int
,@CustomerId int
,@Amount decimal(18,2)
,@TransactionDate datetime
,@TransactionType int
,@TransactionId varchar(50)
,@IsActive bit
,@CreatedDate datetime
,@CreatedByUserId int
,@UpdatedDate datetime
,@UpdatedByUserId int
,@ErrorMessage varchar(3000) output
)
as
begin
If isnull(@Id,0)=0
begin
If @TransactionType = 1
begin
  If @TransactionId != 'Id'
  begin
insert into WalletTransaction(
    CustomerId
	,Amount
	,TransactionDate
	,TransactionType
	,TransactionId
	,IsActive
	,CreatedDate
	,CreatedByUserId
	,UpdatedDate
	,UpdatedByUserId

)
values(
	@CustomerId
	,@Amount
	,@TransactionDate AT TIME ZONE 'UTC' AT TIME ZONE 'India Standard Time'
	,@TransactionType
	,@TransactionId
	,@IsActive
	,@CreatedDate AT TIME ZONE 'UTC' AT TIME ZONE 'India Standard Time'
	,@CreatedByUserId
	,@UpdatedDate AT TIME ZONE 'UTC' AT TIME ZONE 'India Standard Time'
	,@UpdatedByUserId
)
  end
set @ErrorMessage = ''
end
else If @TransactionType = -1
begin
 declare @Credit float,@Debit float,@RemainingAmount float 
set @Credit =(select sum(Amount) from WalletTransaction where CustomerId=@CustomerId and TransactionType=1)
set @Debit =(select isnull(sum(Amount),0) from WalletTransaction where CustomerId=@CustomerId and TransactionType=-1)
set @RemainingAmount = @Credit - @Debit
If @RemainingAmount >0
begin
insert into WalletTransaction(
    CustomerId
	,Amount
	,TransactionDate 
	,TransactionType
	,TransactionId
	,IsActive
	,CreatedDate
	,CreatedByUserId
	,UpdatedDate
	,UpdatedByUserId

)
values(
	@CustomerId
	,@Amount
	,@TransactionDate AT TIME ZONE 'UTC' AT TIME ZONE 'India Standard Time'
	,@TransactionType
	,@TransactionId
	,@IsActive
	,@CreatedDate AT TIME ZONE 'UTC' AT TIME ZONE 'India Standard Time'
	,@CreatedByUserId
	,@UpdatedDate AT TIME ZONE 'UTC' AT TIME ZONE 'India Standard Time'
	,@UpdatedByUserId
)
set @ErrorMessage = ''
end
else
begin
set @ErrorMessage = 'negative balance'
end
end
end
--else
--begin
--update WalletTransaction Set CustomerId=@CustomerId,
--Amount=@Amount,TransactionDate=@TransactionDate,TransactionType=@TransactionType,TransactionId=@TransactionId
--,@IsActive=@IsActive where Id=@Id end
end

GO
```

```sql
/****** Object:  StoredProcedure [dbo].[SP_AutomationSubscription]    Script Date: 27-10-2023 15:01:16 ******/
SET ANSI_NULLS ON
GO
SET QUOTED_IDENTIFIER ON
GO
--exec SP_AutomationSubscription
CREATE proc [dbo].[SP_AutomationSubscription]
as
begin

select CustomerId, sum(Amount) CreditAmount into #CustCredit from WalletTransaction where TransactionType=1 
group by CustomerId
select CustomerId, isnull(sum(Amount),0) DebitAmount into #CustDebit from WalletTransaction where TransactionType=-1 
group by CustomerId

select cc.CustomerId,cc.CreditAmount - isnull(cd.DebitAmount,0) Balance into #CustBalance from #CustCredit cc
left join #CustDebit cd on cc.CustomerId = cd.CustomerId

select * into #Temp from 
(SELECT  sub.CustomerId,sub.CustomerAddressId,sub.ProductId,sub.SubscribeObject
,JSON_VALUE(SubscribeObject,'$.'+LEFT(DATENAME(WEEKDAY, getdate()), 3)) AS Qty,pl.ProductPrice
,(JSON_VALUE(SubscribeObject,'$.'+LEFT(DATENAME(WEEKDAY, getdate()), 3))*pl.ProductPrice) PriceWithOutDiscount
,cast((JSON_VALUE(SubscribeObject,'$.'+LEFT(DATENAME(WEEKDAY, getdate()), 3))*pl.ProductPrice)as decimal(18,2))
-((cast((p.Discount/100) as decimal(18,2)))*(pl.ProductPrice*JSON_VALUE(SubscribeObject,'$.'+LEFT(DATENAME(WEEKDAY, getdate()), 3)))) PriceWithDiscount
,((cast((p.Discount/100) as decimal(18,2)))*(pl.ProductPrice*JSON_VALUE(SubscribeObject,'$.'+LEFT(DATENAME(WEEKDAY, getdate()), 3))))
as DiscountAmount,p.Discount DiscountPer,cb.Balance,st.AvailableStock as AvailableStocks,sub.WareHouseId
FROM Subscription  sub
inner join Product p on p.Id =sub.ProductId
inner join PriceList pl on pl.ProductId = sub.ProductId
inner join #CustBalance cb on cb.CustomerId = sub.CustomerId
--inner join Stock st on st.ProductId = sub.ProductId
inner join WareHouseStock st on st.ProductId = sub.ProductId and st.WareHouseId = sub.WareHouseId
where sub.IsPause=0 and sub.IsActive=1 and pl.EndDate is null 
and CAST(sub.StartDate as Date)<=CAST(getdate() as date))X
where X.Qty<=AvailableStocks

--SELECT  sub.CustomerId,sub.CustomerAddressId,sub.ProductId,sub.SubscribeObject
--,JSON_VALUE(SubscribeObject,'$.'+LEFT(DATENAME(WEEKDAY, getdate()), 3)) AS Qty,pl.ProductPrice
--,(JSON_VALUE(SubscribeObject,'$.'+LEFT(DATENAME(WEEKDAY, getdate()), 3))*pl.ProductPrice) PriceWithOutDiscount
--,cast((JSON_VALUE(SubscribeObject,'$.'+LEFT(DATENAME(WEEKDAY, getdate()), 3))*pl.ProductPrice)as decimal(18,2))
---CEILING((cast((p.Discount/100) as decimal(18,2)))*(pl.ProductPrice*JSON_VALUE(SubscribeObject,'$.'+LEFT(DATENAME(WEEKDAY, getdate()), 3)))) PriceWithDiscount
--,CEILING((cast((p.Discount/100) as decimal(18,2)))*(pl.ProductPrice*JSON_VALUE(SubscribeObject,'$.'+LEFT(DATENAME(WEEKDAY, getdate()), 3))))
--as DiscountAmount,p.Discount DiscountPer,cb.Balance into #Temp
--FROM Subscription  sub
--inner join Product p on p.Id =sub.ProductId
--inner join PriceList pl on pl.ProductId = sub.ProductId
--inner join #CustBalance cb on cb.CustomerId = sub.CustomerId
--where sub.IsPause=0 and sub.IsActive=1 and pl.EndDate is null 
--and CAST(sub.StartDate as Date)<=CAST(getdate() as date) 

DECLARE @CustId int
DECLARE @NoOfProducts int
DECLARE @PriceWithDiscount float
DECLARE @DiscountAmount float
DECLARE @CustomerAddressId int
DECLARE @WareHouseId int
Declare @myresultset Cursor
Set @myresultset = CURSOR for select CustomerId,COUNT(productId) NoOfProducts,
Sum(PriceWithDiscount)+50 PriceWithDiscount,
sum(DiscountAmount) DiscountAmount,CustomerAddressId,WareHouseId
from #Temp where Qty !=0 and PriceWithDiscount<=Balance group by CustomerId,CustomerAddressId,WareHouseId
OPEN @myresultset 
Fetch NEXT from @myresultset
into @CustId,@NoOfProducts,@PriceWithDiscount,@DiscountAmount,@CustomerAddressId,@WareHouseId
While @@Fetch_Status=0
Begin

   DECLARE @dt datetime;
   DECLARE @OutOrderNo varchar(255)
   DECLARE @OutErrorNumber int
   DECLARE @OutErrorMessage varchar(3000)
   DECLARE @OutWalletMessage varchar(3000)
   set @dt= getdate();
   ---Insert Data to Order HDR
 exec SP_AddAutoOrderHdr @Id=null,@OrdDate=@dt,@OrdValue=@PriceWithDiscount,
 @Discount=@DiscountAmount,@DeliveryAddressId=@CustomerAddressId,@SlotId=0,@WareHouseId=@WareHouseId,@IsActive=1,@CreatedDate=@dt,
 @CreatedByUserId=@CustId,@UpdatedDate=@dt,@UpdatedByUserId=@CustId,@IsSubscription=1,@OrderNo = @OutOrderNo output,
 @ErrorNumber = @OutErrorNumber output,@ErrorMessage = @OutErrorMessage output

 exec SP_AddAutoWalletTransaction @Id=null,@CustomerId=@CustId,@Amount=@PriceWithDiscount,@TransactionDate=@dt
,@TransactionType=-1,@TransactionId='',@IsActive=1,@CreatedDate=@dt,@CreatedByUserId=@CustId,@UpdatedDate=@dt
,@UpdatedByUserId=@CustId,@ErrorMessage=@OutWalletMessage output
 

  exec SP_AddAutoOrderPayment @Id=null,@OrderNo=@OutOrderNo,@UserId=@CustId,@PaymentTypeId=2
,@PaymentDate=@dt,@Amount=@PriceWithDiscount,@IsActive=1,@CreatedDate=@dt,@CreatedByUserId=@CustId,@UpdatedDate=@dt
,@UpdatedByUserId=@CustId,@PaymentTransactionId=''
 ---

 ----Insert Data to Order DTL---
DECLARE @ODProductId int
DECLARE @ODQty int
DECLARE @ODProductPrice int
DECLARE @ODDiscountAmount float
DECLARE @ODPriceWithDiscount float
DECLARE @DiscountPer int
Declare @myresultset1 Cursor
Set @myresultset1 = CURSOR for select ProductId,Qty,ProductPrice,DiscountAmount,PriceWithDiscount,DiscountPer
from #Temp where CustomerId = @CustId and Qty !=0 and PriceWithDiscount<=Balance
OPEN @myresultset1 
Fetch NEXT from @myresultset1
into @ODProductId,@ODQty,@ODProductPrice,@ODDiscountAmount,@ODPriceWithDiscount,@DiscountPer
While @@Fetch_Status=0
Begin
   DECLARE @ODdt datetime;
   DECLARE @ODOutErrorNumber int
   DECLARE @ODOutErrorMessage varchar(3000)
   set @ODdt= getdate();
   declare @IV float
   --set @IV = (@ODQty*@ODProductPrice)-@ODDiscountAmount
   --print @ODDiscountAmount
   --print @ODQty*@ODProductPrice
exec SP_AddAutoOrderDtl @Id=null,@OrderNo=@OutOrderNo,@ProductId=@ODProductId,@Qty=@ODQty,
@Price = @ODProductPrice,@Discount = @DiscountPer,@ItemValue =@ODPriceWithDiscount,
@IsActive=1,@CreatedDate=@ODdt,@CreatedByUserId=@CustId,@UpdatedDate=@ODdt,@UpdatedByUserId=@CustId,
@DiscountedAmt = @ODDiscountAmount,@ErrorNumber = @ODOutErrorNumber output,
@ErrorMessage = @ODOutErrorMessage output


  
 Fetch Next from @myresultset1
 into @ODProductId,@ODQty,@ODProductPrice,@ODDiscountAmount,@ODPriceWithDiscount,@DiscountPer
End
close @myresultset1

 ------------------------

 Fetch Next from @myresultset
 into @CustId,@NoOfProducts,@PriceWithDiscount,@DiscountAmount,@CustomerAddressId,@WareHouseId
End
close @myresultset

drop table #Temp
drop table #CustCredit
drop table #CustDebit
drop table #CustBalance
end
GO
```

```sql
/****** Object:  StoredProcedure [dbo].[SP_GetOrdersDetails]    Script Date: 27-10-2023 15:01:16 ******/
SET ANSI_NULLS ON
GO
SET QUOTED_IDENTIFIER ON
GO
CREATE proc [dbo].[SP_GetOrdersDetails](@StartDate date,@EndDate date,@WareHouseId Int)
as
begin

select X.Status,sum(X.NoOfOrders) NoOfOrders from
((select oh.Status,count(oh.OrderNo) NoOfOrders from OrderHdr oh

where oh.WareHouseId = ISNULL(@WareHouseId, oh.WareHouseId) and  cast(oh.OrdDate as Date)>=@StartDate 
and  cast(oh.OrdDate as Date)<=@EndDate and oh.Status not in ('Confirmed') group by  GROUPING SETS (
        (oh.Status),
        ()))
        union
select 'Cancel' Status,0 NoOfOrders
union
select null Status,0 NoOfOrders
union
select 'Placed' Status,0 NoOfOrders
union
select 'Delivered' Status,0 NoOfOrders
union
select 'Preparing for Dispatch' Status,0 NoOfOrders
union
select 'On the way' Status,0 NoOfOrders)X group by Status

end
GO
```

# [dbo].[SP_GetWalletBalance]

```sql
/****** Object:  StoredProcedure [dbo].[SP_GetWalletBalance]    Script Date: 27-10-2023 15:01:16 ******/
SET ANSI_NULLS ON
GO
SET QUOTED_IDENTIFIER ON
GO
--exec SP_GetWalletBalance 17
CREATE Proc [dbo].[SP_GetWalletBalance](@CustomerId int)
as
begin
declare @Credit float,@Debit float
set @Credit =(select sum(Amount) from WalletTransaction where CustomerId=@CustomerId and TransactionType=1)

set @Debit =(select isnull(sum(Amount),0) from WalletTransaction where CustomerId=@CustomerId and TransactionType=-1)

select cast(@Credit - @Debit as decimal(18,2)) balance
end
GO
```

# [dbo].[SP_PushNotificationForWalletBalace]

```sql
/****** Object:  StoredProcedure [dbo].[SP_PushNotificationForWalletBalace]    Script Date: 27-10-2023 15:01:16 ******/
SET ANSI_NULLS ON
GO
SET QUOTED_IDENTIFIER ON
GO

--exec SP_PushNotificationForWalletBalace
CREATE Proc [dbo].[SP_PushNotificationForWalletBalace]
as
begin
select CustomerId, sum(Amount) CreditAmount into #CustCredit from WalletTransaction where TransactionType=1 
group by CustomerId
select CustomerId, isnull(sum(Amount),0) DebitAmount into #CustDebit from WalletTransaction where TransactionType=-1 
group by CustomerId

select cc.CustomerId,cc.CreditAmount - cd.DebitAmount Balance into #CustBalance from #CustCredit cc
left join #CustDebit cd on cc.CustomerId = cd.CustomerId

SELECT  sub.CustomerId,sub.CustomerAddressId,sub.ProductId,sub.SubscribeObject
,JSON_VALUE(SubscribeObject,'$.'+LEFT(DATENAME(WEEKDAY, getdate()), 3)) AS Qty,pl.ProductPrice
,(JSON_VALUE(SubscribeObject,'$.'+LEFT(DATENAME(WEEKDAY, getdate()), 3))*pl.ProductPrice) PriceWithOutDiscount
,cast((JSON_VALUE(SubscribeObject,'$.'+LEFT(DATENAME(WEEKDAY, getdate()), 3))*pl.ProductPrice)as decimal(18,2))
-CEILING((cast((p.Discount/100) as decimal(18,2)))*(pl.ProductPrice*JSON_VALUE(SubscribeObject,'$.'+LEFT(DATENAME(WEEKDAY, getdate()), 3)))) PriceWithDiscount
,CEILING((cast((p.Discount/100) as decimal(18,2)))*(pl.ProductPrice*JSON_VALUE(SubscribeObject,'$.'+LEFT(DATENAME(WEEKDAY, getdate()), 3))))
as DiscountAmount,p.Discount DiscountPer,cb.Balance into #Temp
FROM Subscription  sub
inner join Product p on p.Id =sub.ProductId
inner join PriceList pl on pl.ProductId = sub.ProductId
inner join #CustBalance cb on cb.CustomerId = sub.CustomerId
where sub.IsPause=0 and sub.IsActive=1 and pl.EndDate is null 
and CAST(sub.StartDate as Date)<=CAST(getdate() as date) 

DECLARE @CustId int
DECLARE @NoOfRecords int
Declare @myresultset Cursor
Create table #PushNotification(CustomerId int,Message varchar(250),CreatedDate datetime)
Set @myresultset = CURSOR for select CustomerId,COUNT(*) NoOfRecords
from #Temp where Qty !=0 and PriceWithDiscount>Balance group by CustomerId
OPEN @myresultset 
Fetch NEXT from @myresultset
into @CustId,@NoOfRecords
While @@Fetch_Status=0
Begin
  insert into #PushNotification(CustomerId,Message,CreatedDate)
  values(@CustId,'Your Wallet Balance is Low',getdate())
 Fetch Next from @myresultset
 into @CustId,@NoOfRecords
End
close @myresultset
select * from #PushNotification
drop table #Temp
drop table #CustCredit
drop table #CustDebit
drop table #CustBalance
drop table #PushNotification
end
GO
```

# [dbo].[SP_UpdateOrderStatus]

```sql
/****** Object:  StoredProcedure [dbo].[SP_UpdateOrderStatus]    Script Date: 27-10-2023 15:01:16 ******/
SET ANSI_NULLS ON
GO
SET QUOTED_IDENTIFIER ON
GO
CREATE Proc [dbo].[SP_UpdateOrderStatus](@OrderNo varchar(50),@OrderStatus varchar(50),@UserId int,@ReasonForCancel Varchar(255))
as
begin
  Update OrderHdr set Status=@OrderStatus ,ReasonForCancel = @ReasonForCancel  where OrderNo=@OrderNo and CreatedByUserId=@UserId
  insert OrderNotification (OrderNo,Status,CreatedDate,CustomerId) values (@OrderNo,@OrderStatus,GETDATE(),@UserId)

  -- if Order Cancelled then add Order Amount to Wallet 
  if @OrderStatus = 'Cancel'
  begin
    declare @RefundAmount float 
	 DECLARE @dt datetime;
     DECLARE @OutWalletMessage varchar(3000)
     set @dt= getdate();
	set @RefundAmount = (select OrdValue from OrderHdr where OrderNo=@OrderNo and CreatedByUserId=@UserId);
	 exec SP_AddAutoWalletTransaction @Id=null,@CustomerId=@UserId,@Amount=@RefundAmount,@TransactionDate=@dt
,@TransactionType=1,@TransactionId='Refund',@IsActive=1,@CreatedDate=@dt,@CreatedByUserId=@UserId,@UpdatedDate=@dt
,@UpdatedByUserId=@UserId,@ErrorMessage=@OutWalletMessage output

  end

  if @OrderStatus = 'On the way'
  begin
DECLARE @ProductId int
DECLARE @Qty int
Declare @myresultset Cursor
Set @myresultset = CURSOR for select ProductId,Qty from OrderDtl where OrderNo in
(select OrderNo from OrderHdr where Status='On the way' and OrderNo=@OrderNo)
OPEN @myresultset 
Fetch NEXT from @myresultset
into @ProductId,@Qty
While @@Fetch_Status=0
Begin
  update Stock set TotalStocks=TotalStocks-@Qty where ProductId=@ProductId
 Fetch Next from @myresultset
 into @ProductId,@Qty
End
close @myresultset
 end
  else if @OrderStatus = 'Cancel'
  begin
 DECLARE @ProductId1 int
DECLARE @Qty1 int
Declare @myresultset1 Cursor
Set @myresultset1 = CURSOR for select ProductId,Qty from OrderDtl where OrderNo in
(select OrderNo from OrderHdr where Status='Cancel' and OrderNo=@OrderNo)
OPEN @myresultset1 
Fetch NEXT from @myresultset1
into @ProductId1,@Qty1
While @@Fetch_Status=0
Begin
  update Stock set AvailableStocks=AvailableStocks+@Qty1 where ProductId=@ProductId1


 Fetch Next from @myresultset1
 into @ProductId1,@Qty1
End
close @myresultset1
  end
end
GO
```
