```sql
/****** Object:  StoredProcedure [dbo].[SP_AddCategory]    Script Date: 27-10-2023 15:02:48 ******/
SET ANSI_NULLS ON
GO
SET QUOTED_IDENTIFIER ON
GO
CREATE Procedure [dbo].[SP_AddCategory]
(@Id int,
@Code varchar(50),
@Name varchar(100),
@IsActive bit,
@CreatedByUserId int,
@CreatedDate datetime,
@UpdatedByUserId int,
@UpdatedDate datetime,
@ErrorNumber INT OUTPUT
,@ErrorMessage NVARCHAR(4000) OUTPUT
)
AS
BEGIN
	BEGIN TRANSACTION

	BEGIN TRY
IF ISNULL (@Id,0)=0
begin
Insert into Category(
Code,
Name,
IsActive,
CreatedByUserId,
CreatedDate,
UpdatedByUserId,
UpdatedDate
)
values(
@Code,
@Name,
@IsActive,
@CreatedByUserId,
@CreatedDate,
@UpdatedByUserId,
@UpdatedDate
)
End
ELSE
Update Category Set Code=@Code,Name=@Name,IsActive=@IsActive,
UpdatedByUserId=@UpdatedByUserId,[UpdatedDate] = ISNULL(@UpdatedDate, GETDATE()) AT TIME ZONE 'UTC' AT TIME ZONE 'India Standard Time' where Id=@Id
END TRY

	BEGIN CATCH
		SELECT @ErrorNumber = ERROR_NUMBER()
			,@ErrorMessage = ERROR_MESSAGE()

		IF @@TRANCOUNT > 0
			ROLLBACK TRANSACTION;
		RAISERROR ('Error: %d: %s', 16, 1, @ErrorNumber, @ErrorMessage) 
	END CATCH;

	IF @@TRANCOUNT > 0
		COMMIT TRANSACTION;

	SET @ErrorNumber = 0
	SET @ErrorMessage = 'Success'
END




GO
```

```sql
/****** Object:  StoredProcedure [dbo].[SP_AddPayment]    Script Date: 27-10-2023 15:02:48 ******/
SET ANSI_NULLS ON
GO
SET QUOTED_IDENTIFIER ON
GO
CREATE PROCEDURE [dbo].[SP_AddPayment] 
(
 @Id int,
 @InvoiceNo Int,
 @PONumber Int
 ,@PaymentMode Int
 ,@PaymentTerms Int
,@TransactionID varchar(50)
,@TransactionAmount Decimal(18,3)
,@TransactionDate Datetime
,@TransactionImg Varchar(255)
,@Remark Varchar(255)
,@AmountDue int
,@AmountPaid int
,@IsActive bit
,@CreatedDate datetime
,@CreatedByUserId int
,@UpdatedDate datetime
,@UpdatedByUserId int
,@ErrorNumber INT OUTPUT
,@ErrorMessage NVARCHAR(4000) OUTPUT
)
AS
BEGIN
	BEGIN TRANSACTION

	BEGIN TRY
		IF ISNULL(@Id,0)=0
			INSERT INTO [dbo].[PaymentConfirmation] (
				 [InvoiceNo]
				,[PONumber]
				,[PaymentMode]
				,[PaymentTerms]
				,[TransactionID]
				,[TransactionAmount]
				,[TransactionDate]
				,[TransactionImg]
				,AmountDue
                ,AmountPaid 
				,[Remark]
				,[IsActive]
				,[CreatedDate]
				,[CreatedByUserId]
				,[UpdatedDate]
				,[UpdatedByUserId]
				
				)
			VALUES (
				 @InvoiceNo
				,@PONumber
				,@PaymentMode
				,@PaymentTerms
				,@TransactionID
				,@TransactionAmount
				,@TransactionDate
				,@TransactionImg
				,@AmountDue - @AmountPaid 
                ,@AmountPaid 
				,@Remark
				,ISNULL(@IsActive, 1)
				,ISNULL(@CreatedDate, GETDATE())
				,@CreatedByUserId
				,ISNULL(@UpdatedDate, GETDATE())
				,@UpdatedByUserId
				
				)
		ELSE
			UPDATE [dbo].[PaymentConfirmation]
			SET  [InvoiceNo] = @InvoiceNo
				,[PONumber] =  @PONumber
				,[PaymentMode] = @PaymentMode
				,[PaymentTerms] = @PaymentTerms
				,[TransactionID] = @TransactionID
                ,[TransactionAmount] = @TransactionAmount
				,[TransactionDate] = @TransactionDate
				,[TransactionImg] = @TransactionImg
				,[AmountPaid]=@AmountPaid
				,[AmountDue]=AmountDue-@AmountPaid	
				
				,[Remark] = @Remark
				,[IsActive] = ISNULL(@IsActive, 1)
				,[UpdatedDate] = ISNULL(@UpdatedDate, GETDATE()) AT TIME ZONE 'UTC' AT TIME ZONE 'India Standard Time'
				,[UpdatedByUserId] = @UpdatedByUserId
				
			WHERE Id = @Id
	END TRY

	BEGIN CATCH
		SELECT @ErrorNumber = ERROR_NUMBER()
			,@ErrorMessage = ERROR_MESSAGE()

		IF @@TRANCOUNT > 0
			ROLLBACK TRANSACTION;
		RAISERROR ('Error: %d: %s', 16, 1, @ErrorNumber, @ErrorMessage) 
	END CATCH;

	IF @@TRANCOUNT > 0
		COMMIT TRANSACTION;

	SET @ErrorNumber = 0
	SET @ErrorMessage = 'Success'
END
GO
```

```sql
/****** Object:  StoredProcedure [dbo].[SP_ApprovalPOHdrDetails]    Script Date: 27-10-2023 15:02:48 ******/
SET ANSI_NULLS ON
GO
SET QUOTED_IDENTIFIER ON
GO
CREATE procedure [dbo].[SP_ApprovalPOHdrDetails]
as begin
SELECT 
		 t.[Id]
		,t.[PONo]
		,u.[BusinessName]
		,t.[BusinessName] as partnerId
		,c.[Name] as POStatus
		,t.[POStatus] as statusId
		,t.[PartnerGSTINNo]
		,t.[ShippingAddress]
		,t.[BillingAddress]
		,t.[RequireDate]
		,k.Name As PaymentTerms
		,t.PaymentTerms as PaymentTermsId
		,l.Name as ModeofDelivery
		,t.ModeofDelivery as ModeofDeliveryId
		,w.Name as Modeofpayment
		,t.Modeofpayment as ModeofpaymentId
        ,t.[IsActive]
		,t.[CreatedDate]
		,t.[CreatedByUserId]
		,t.[UpdatedDate]
		,t.[UpdatedByUserId]
		,pod.GrandTotal
		
	FROM [dbo].[POHdr] t 
	INNER JOIN LookupDtl c ON c.Id=t.POStatus
	INNER JOIN LookupDtl k ON k.Id=t.PaymentTerms
	INNER JOIN LookupDtl l ON l.Id=t.ModeofDelivery
	INNER JOIN LookupDtl w ON w.Id=t.Modeofpayment
	INNER JOIN OnBoardPartner u ON u.Id=t.BusinessName
	inner Join (SELECT POHdrId, SUM(TaxableValue) as GrandTotal FROM PODtl Group By POHdrId) pod On  pod.POHdrId= t.Id
	AND t.IsActive = 1 and t.POStatus=3
	union
		select 0 [Id]
		,'SPO00001' [PONo]
		,'' [BusinessName]
		,0 partnerId
		,'Accepted' POStatus
		,3 as statusId
		,null [PartnerGSTINNo]
		,null [ShippingAddress]
		,null [BillingAddress]
		,getdate() [RequireDate]
		,'90 Days' As PaymentTerms
		,28 as PaymentTermsId
		,'Transport' as ModeofDelivery
		,6 as ModeofDeliveryId
		,'Online' as Modeofpayment
		,29 as ModeofpaymentId
        ,1 IsActive
		,getdate() [CreatedDate]
		,1 [CreatedByUserId]
		,getdate() [UpdatedDate]
		,1 [UpdatedByUserId]
		,0 GrandTotal
end
GO
```

# [dbo].[SP_GetUserAccessRights]

# [dbo].[SP_RequestedPOHdrDetails]

```sql
/****** Object:  StoredProcedure [dbo].[SP_RequestedPOHdrDetails]    Script Date: 27-10-2023 15:02:48 ******/
SET ANSI_NULLS ON
GO
SET QUOTED_IDENTIFIER ON
GO
CREATE procedure [dbo].[SP_RequestedPOHdrDetails]
as begin
SELECT 
		 t.[Id]
		,t.[PONo]
		,u.[BusinessName]
		,t.[BusinessName] as partnerId
		,c.[Name] as POStatus
		,t.[POStatus] as statusId
		,t.[PartnerGSTINNo]
		,t.[ShippingAddress]
		,t.[BillingAddress]
		,t.[RequireDate]
		,k.Name As PaymentTerms
		,t.PaymentTerms as PaymentTermsId
		,l.Name as ModeofDelivery
		,t.ModeofDelivery as ModeofDeliveryId
		,w.Name as Modeofpayment
		,t.Modeofpayment as ModeofpaymentId
        ,t.[IsActive]
		,t.[CreatedDate]
		,t.[CreatedByUserId]
		,t.[UpdatedDate]
		,t.[UpdatedByUserId]
		,pod.GrandTotal
		
	FROM [dbo].[POHdr] t 
	INNER JOIN LookupDtl c ON c.Id=t.POStatus
	INNER JOIN LookupDtl k ON k.Id=t.PaymentTerms
	INNER JOIN LookupDtl l ON l.Id=t.ModeofDelivery
	INNER JOIN LookupDtl w ON w.Id=t.Modeofpayment
	INNER JOIN OnBoardPartner u ON u.Id=t.BusinessName
	inner Join (SELECT POHdrId, SUM(TaxableValue) as GrandTotal FROM PODtl Group By POHdrId) pod On  pod.POHdrId= t.Id
	
		AND t.IsActive = 1 and POStatus=1
end
GO
```
