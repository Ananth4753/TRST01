# [dbo].[SP_AddUsers]

```sql
/****** Object:  StoredProcedure [dbo].[SP_AddUsers]    Script Date: 27-10-2023 15:16:00 ******/
SET ANSI_NULLS ON
GO
SET QUOTED_IDENTIFIER ON
GO
CREATE PROCEDURE [dbo].[SP_AddUsers] 
(
 @Id int
,@UserName varchar(30)
,@Password nvarchar(1024)
,@FirstName varchar(50)
,@MiddileName varchar(50)
,@LastName varchar(50)
,@EmailId varchar(100)
,@MobileNumber bigint
,@RoleId int
,@ViewYN bit
,@CreatedDate datetime
,@CreatedByUserId int
,@UpdatedDate datetime
,@UpdatedByUserId int
,@ErrorNumber INT OUTPUT
,@ErrorMessage NVARCHAR(4000) OUTPUT
)
AS
BEGIN
	BEGIN TRANSACTION

	BEGIN TRY
		IF ISNULL(@Id,0)=0
			INSERT INTO [dbo].[Users] (
				 [UserName]
				,[Password]
				,[FirstName]
				,[MiddileName]
				,[LastName]
				,[EmailId]
				,[MobileNumber]
				,[RoleId]
				,[ViewYN]
				,[CreatedDate]
				,[CreatedByUserId]
				,[UpdatedDate]
				,[UpdatedByUserId]
				)
			VALUES (
				@UserName
				,@Password
				,@FirstName
				,@MiddileName
				,@LastName
				,@EmailId
				,@MobileNumber
				,@RoleId
				,ISNULL(@ViewYN, 1)
				,ISNULL(@CreatedDate, GETDATE())
				,@CreatedByUserId
				,ISNULL(@UpdatedDate, GETDATE())
				,@UpdatedByUserId
				)
		ELSE
			UPDATE [dbo].[Users]
			SET  
				[UserName] = @UserName
				,[Password] = @Password
				,[FirstName] = @FirstName
				,[MiddileName] = @MiddileName
				,[LastName] = @LastName
				,[EmailId] = @EmailId
				,[MobileNumber] = @MobileNumber
				,[RoleId] = @RoleId
				,[ViewYN] = ISNULL(@ViewYN, 1)
				,[UpdatedDate] = ISNULL(@UpdatedDate, GETDATE())
				,[UpdatedByUserId] = @UpdatedByUserId
			WHERE Id = @Id
	END TRY

	BEGIN CATCH
		SELECT @ErrorNumber = ERROR_NUMBER()
			,@ErrorMessage = ERROR_MESSAGE()

		IF @@TRANCOUNT > 0
			ROLLBACK TRANSACTION;
		RAISERROR ('Error: %d: %s', 16, 1, @ErrorNumber, @ErrorMessage) 
	END CATCH;

	IF @@TRANCOUNT > 0
		COMMIT TRANSACTION;

	SET @ErrorNumber = 0
	SET @ErrorMessage = 'Success'
END
GO
```

# [dbo].[SP_GetConfigurationDtlByConfigHdrId]

```sql
/****** Object:  StoredProcedure [dbo].[SP_GetConfigurationDtlByConfigHdrId]    Script Date: 27-10-2023 15:16:00 ******/
SET ANSI_NULLS ON
GO
SET QUOTED_IDENTIFIER ON
GO
CREATE proc [dbo].[SP_GetConfigurationDtlByConfigHdrId](@P_ConfigurationHdrId int)
as
begin
select * from
(select u.UniversityName,b.Name BranchName,lud.Year,cast(cast(lud.Year as varchar(10))+'-'+cast(lud.Semester as varchar(10)) as varchar(255)) YearSem,x.Subjects from
(SELECT ConfigurationHdrId,BranchId,YearId ,(select STRING_AGG(Subject, '***') from Subjects where Id in (select Item from [SplitString]((STRING_AGG(SubjectId, ', ')),',') )  
) Subjects
FROM ConfigurationDtl where ConfigurationHdrId =@P_ConfigurationHdrId
GROUP BY ConfigurationHdrId,BranchId,YearId)x
inner join Branch b on b.Id = x.BranchId
inner join RegulationDtl lud on lud.Id = x.YearId
inner join ConfigurationHdr chd on chd.Id = x.ConfigurationHdrId
inner join University u on u.Id = chd.UniversityId)y
order by y.BranchName,y.YearSem
end
GO
```

# [dbo].[SP_GetMenuDetailsByUserId]

```sql
/****** Object:  StoredProcedure [dbo].[SP_GetMenuDetailsByUserId]    Script Date: 27-10-2023 15:16:00 ******/
SET ANSI_NULLS ON
GO
SET QUOTED_IDENTIFIER ON
GO
CREATE PROCEDURE [dbo].[SP_GetMenuDetailsByUserId] (@P_UserId INT)
AS
BEGIN
	SELECT RM.Id,
	RM.RoleId
		,R.[Name] [Role]
		,RM.MenuId
		,CASE WHEN m.Path LIKE'http%' THEN 'REPORTS' ELSE  PM.Name END ParentMenu
		,Case when m.MenuType='M' then 'Menu' when m.MenuType='S' then 'Sub Menu' else 'Item' end MenuType
		,M.[Name] [Menu]
		,m.Path
		,m.Icon
	FROM dbo.RoleMenu RM
	INNER JOIN Roles R ON RM.RoleId = R.Id
	INNER JOIN Menu M ON RM.MenuId = M.Id
	LEFT JOIN Menu PM ON M.ParentMenuId = PM.Id
	INNER JOIN Users u ON u.RoleId=r.Id
	WHERE u.Id = ISNULL(@P_UserId, u.Id)
	AND RM.ViewYN = 1
	AND ISNULL(m.Path,'')!=''
	AND m.Path not like 'http%'
	ORDER By r.name,CASE WHEN m.Path LIKE'http%' THEN 9999 ELSE  PM.[Order] END ,m.[Order]
END
GO
```
