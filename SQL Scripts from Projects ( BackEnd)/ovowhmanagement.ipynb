# [dbo].[SP_UpdateProductLevelStockIn]

# [dbo].[SP_ValidateUser]

```sql
/****** Object:  StoredProcedure [dbo].[SP_ValidateUser]    Script Date: 27-10-2023 15:21:37 ******/
SET ANSI_NULLS ON
GO
SET QUOTED_IDENTIFIER ON
GO

CREATE PROCEDURE [dbo].[SP_ValidateUser]
	@UserName VARCHAR(50),
	@Password VARCHAR(50)    
AS

BEGIN
	SET NOCOUNT ON;
		SELECT 
			u.Id,
			u.FirstName,
			u.LastName,
			u.UserName,
			u.MobileNo,
			u.Password
			,u.WareHouseId
		,w.Name as WareHouseName
		FROM Users u 
		left join WareHouse w on w.Id = u.WareHouseId
		WHERE u.UserName=@UserName AND u.Password=@Password AND u.ViewYN=1
END
GO
```
