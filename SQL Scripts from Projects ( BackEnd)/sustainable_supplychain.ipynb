```sql
/****** Object:  StoredProcedure [dbo].[SP_AddDealer]    Script Date: 27-10-2023 15:11:24 ******/
SET ANSI_NULLS ON
GO
SET QUOTED_IDENTIFIER ON
GO
CREATE procedure [dbo].[SP_AddDealer](
@Id int,
@Name varchar(255),
@PrimaryContactNo bigint,
@Address varchar(500),
@Village int,
@Latitude decimal(12,9),
@Longitute decimal(12,9),
@IsActive bit,
@CreatedDate datetime,
@CreatedByUserId int,
@UpdatedDate datetime,
@UpdatedByUserId int,
@ErrorNumber INT OUTPUT,
@ErrorMessage NVARCHAR(4000) OUTPUT
)
AS
BEGIN

    BEGIN TRANSACTION

	BEGIN TRY
	   IF ISNULL(@Id,0) = 0
	   INSERT INTO Dealer(
	   Name,
	   PrimaryContactNo,
	   Address,
	   Village,
	   Latitude,
	   Longitute,
	   IsActive,
	   CreatedDate,
	   CreatedByUserId,
	   UpdatedDate,
	   UpdatedByUserId
	   )
	   VALUES(
	    @Name,
	   @PrimaryContactNo,
	   @Address,
	   @Village,
	   @Latitude,
	   @Longitute,
	   @IsActive,
	   @CreatedDate,
	   @CreatedByUserId,
	   @UpdatedDate,
	   @UpdatedByUserId
	   )
      ELSE 
	      UPDATE Dealer 
   SET  Name = @Name,
		PrimaryContactNo = @PrimaryContactNo,
		Address = @Address,
		Village = @Village,
		Latitude = @Latitude,
		Longitute = @Longitute,
		IsActive = @IsActive,
		UpdatedDate = @UpdatedDate,
		UpdatedByUserId = @UpdatedByUserId
		WHERE Id = @Id
	END TRY
	    BEGIN CATCH
		 SELECT @ErrorNumber = ERROR_NUMBER(),
		        @ErrorMessage = ERROR_MESSAGE()

				IF @@TRANCOUNT > 0
				ROLLBACK TRANSACTION;
				RAISERROR ('Error: %d: %s', 16, 1, @ErrorNumber, @ErrorMessage) 

		END CATCH
		IF @@TRANCOUNT > 0
		COMMIT TRANSACTION;

	SET @ErrorNumber = 0
	SET @ErrorMessage = 'Success'
END
GO
```

# [dbo].[SP_GetMenuDetailsByUserId]

```sql
/****** Object:  StoredProcedure [dbo].[SP_GetUserAccessRights]    Script Date: 27-10-2023 15:11:24 ******/
SET ANSI_NULLS ON
GO
SET QUOTED_IDENTIFIER ON
GO
CREATE PROCEDURE [dbo].[SP_GetUserAccessRights](@UserId INT,@Path VARCHAR(128)) AS
BEGIN

SELECT TOP 1 
rm.AddYN,rm.EditYN,rm.DeleteYN,rm.ViewYN
FROM RoleMenu rm 
INNER JOIN Menu m ON rm.MenuId = m.Id
INNER JOIN Users u ON u.RoleId = rm.RoleId
WHERE u.Id=@UserId
AND m.MenuType='I'
AND m.Path = @Path
END
GO
```

# [dbo].[SP_GetUserDetails]

```sql
/****** Object:  StoredProcedure [dbo].[SP_GetUserDetails]    Script Date: 27-10-2023 15:11:24 ******/
SET ANSI_NULLS ON
GO
SET QUOTED_IDENTIFIER ON
GO
CREATE PROCEDURE [dbo].[SP_GetUserDetails] (@Id INT)
AS
BEGIN
	SELECT 
		 t.[Id]
		,t.[UserName]
		,t.[Password]
		,t.[FirstName]
		,t.[LastName]
		,t.[EmailId]
		,t.[MobileNumber]
		,t.[RoleId]
		,r.Name AS RoleName
		,t.[IsActive]
		,t.[CreatedDate]
		,c.Name AS CreatedByUser
		,t.CreatedByUserId
		,t.[UpdatedDate]
		,u.Name AS UpdatedByUser
		,t.UpdatedByUserId
	FROM [dbo].[Users] t
	INNER JOIN VW_UserInfo c ON c.Id=t.CreatedByUserId
	INNER JOIN VW_UserInfo u ON u.Id=t.UpdatedByUserId
	INNER JOIN Roles r on r.Id = t.RoleId
	WHERE t.Id = ISNULL(@Id, t.Id) order by CreatedDate desc
		--AND t.IsActive = 1

END
GO
```

# [dbo].[SP_ValidateUser]

```sql
/****** Object:  StoredProcedure [dbo].[SP_ValidateUser]    Script Date: 27-10-2023 15:11:24 ******/
SET ANSI_NULLS ON
GO
SET QUOTED_IDENTIFIER ON
GO
--exec SP_ValidateUser 'DeveloperUser','b56e0b4ea4962283bee762525c2d490f'
CREATE PROCEDURE [dbo].[SP_ValidateUser]
	@UserName VARCHAR(255),
	@Password VARCHAR(255)    
AS

BEGIN
	
		SELECT 
		 t.[Id]
		,[UserName]
		,[Password]
		,[FirstName]
		,[LastName]
		,[EmailId]
		,[MobileNumber]
		,[RoleId]
		,t.[IsActive]
		,t.[CreatedDate]
		,[CreatedByUserId]
		,t.[UpdatedDate]
		,[UpdatedByUserId]
		,r.Name as RoleName
	FROM [dbo].[Users] t
	inner join [dbo].[Roles] r on r.Id = t.RoleId
	WHERE t.UserName=@UserName AND t.Password=@Password AND t.IsActive=1
END
GO
```
