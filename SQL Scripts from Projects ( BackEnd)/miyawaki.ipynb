# [dbo].[SP_GetMenuDetailsByUserId]

```sql
/****** Object:  StoredProcedure [dbo].[SP_GetMenuDetailsByUserId]    Script Date: 27-10-2023 15:09:25 ******/
SET ANSI_NULLS ON
GO
SET QUOTED_IDENTIFIER ON
GO
CREATE PROCEDURE [dbo].[SP_GetMenuDetailsByUserId] (@P_UserId INT)
AS
BEGIN
	SELECT RM.Id,
	RM.RoleId
		,R.[Name] [Role]
		,RM.MenuId
		,CASE WHEN m.Path LIKE'http%' THEN 'REPORTS' ELSE  PM.Name END ParentMenu
		,Case when m.MenuType='M' then 'Menu' when m.MenuType='S' then 'Sub Menu' else 'Item' end MenuType
		,M.[Name] [Menu]
		,m.Path
		,m.Icon
	FROM dbo.RoleMenu RM
	INNER JOIN Roles R ON RM.RoleId = R.Id
	INNER JOIN Menu M ON RM.MenuId = M.Id
	LEFT JOIN Menu PM ON M.ParentMenuId = PM.Id
	INNER JOIN Users u ON u.RoleId=r.Id
	WHERE u.Id = ISNULL(@P_UserId, u.Id)
	AND RM.ViewYN = 1
	AND ISNULL(m.Path,'')!=''
	AND m.Path not like 'http%'
	ORDER By r.name,CASE WHEN m.Path LIKE'http%' THEN 9999 ELSE  PM.[Order] END ,m.[Order]
END
GO
```

# [dbo].[SP_ValidateUser]

```sql
/****** Object:  StoredProcedure [dbo].[SP_ValidateUser]    Script Date: 27-10-2023 15:09:25 ******/
SET ANSI_NULLS ON
GO
SET QUOTED_IDENTIFIER ON
GO
CREATE PROCEDURE [dbo].[SP_ValidateUser]
	@UserName VARCHAR(50),
	@Password VARCHAR(50)   
AS

BEGIN
	SET NOCOUNT ON;
		SELECT 
			u.Id,
			u.FirstName,
			u.LastName,
			u.UserName,
			u.MobileNumber,
			u.Password,
			u.OrgId
		FROM Users u 
		WHERE u.UserName=@UserName AND u.Password=@Password AND u.ViewYN=1
END
GO
```
