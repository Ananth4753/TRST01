# [dbo].[SP_AddMenu]

```sql
/****** Object:  StoredProcedure [dbo].[SP_AddMenu]    Script Date: 27-10-2023 15:08:31 ******/
SET ANSI_NULLS ON
GO
SET QUOTED_IDENTIFIER ON
GO
CREATE PROCEDURE [dbo].[SP_AddMenu] 
(
 @Id int = NULL
,@Code varchar(50)
,@Name varchar(100)
,@MenuType varchar(1)
,@ParentMenuId int
,@ViewYN bit
,@CreatedDate datetime
,@CreatedByUserId int
,@UpdatedDate datetime
,@UpdatedByUserId int
,@Path varchar(1024)
,@Icon varchar(1024)
,@Order int
,@ErrorNumber INT OUTPUT
,@ErrorMessage NVARCHAR(4000) OUTPUT
)
AS
BEGIN
	BEGIN TRANSACTION

	BEGIN TRY
		IF ISNULL(@Id,0)=0
			INSERT INTO [dbo].[Menu] (
				 [Code]
				,[Name]
				,[MenuType]
				,[ParentMenuId]
				,[ViewYN]
				,[CreatedDate]
				,[CreatedByUserId]
				,[UpdatedDate]
				,[UpdatedByUserId]
				,[Path]
				,[Icon]
				,[Order]
				)
			VALUES (
				 @Code
				,@Name
				,@MenuType
				,@ParentMenuId
				,ISNULL(@ViewYN, 1)
				,ISNULL(@CreatedDate, GETDATE())
				,@CreatedByUserId
				,ISNULL(@UpdatedDate, GETDATE())
				,@UpdatedByUserId
				,@Path
				,@Icon
				,@Order
				)
		ELSE
			UPDATE [dbo].[Menu]
			SET  [Code] = @Code
				,[Name] = @Name
				,[MenuType] = @MenuType
				,[ParentMenuId] = @ParentMenuId
				,[ViewYN] = ISNULL(@ViewYN, 1)
				,[UpdatedDate] = ISNULL(@UpdatedDate, GETDATE())
				,[UpdatedByUserId] = @UpdatedByUserId
				,[Path] = @Path
				,[Icon] = @Icon
				,[Order] = @Order
			WHERE Id = @Id
	END TRY

	BEGIN CATCH
		SELECT @ErrorNumber = ERROR_NUMBER()
			,@ErrorMessage = ERROR_MESSAGE()

		IF @@TRANCOUNT > 0
			ROLLBACK TRANSACTION;
		RAISERROR ('Error: %d: %s', 16, 1, @ErrorNumber, @ErrorMessage) 
	END CATCH;

	IF @@TRANCOUNT > 0
		COMMIT TRANSACTION;

	SET @ErrorNumber = 0
	SET @ErrorMessage = 'Success'
END
GO
```

# [dbo].[SP_GetMenuDetails]

```sql
/****** Object:  StoredProcedure [dbo].[SP_GetMenuDetails]    Script Date: 27-10-2023 15:08:31 ******/
SET ANSI_NULLS ON
GO
SET QUOTED_IDENTIFIER ON
GO
CREATE PROCEDURE [dbo].[SP_GetMenuDetails] (@P_Id INT)
AS
BEGIN
	SELECT m.Id
		,m.Code
		,m.Name
		,Case when m.MenuType='M' then 'Menu' when m.MenuType='S' then 'Sub Menu' else 'Item' end MenuType
		,mm.Name as ParentMenu
		,m.[Path]
		,m.[Icon]
		,m.[Order]
		,m.ViewYN
		,m.CreatedDate
		,u.Name as 'CreatedByUser'
		,m.UpdatedDate
		,um.Name as 'UpdatedByUser'
	FROM dbo.Menu m 
	inner join UserInfo u on m.CreatedByUserId=u.Id 
	left join menu mm on mm.Id=m.ParentMenuId
	inner join UserInfo Um ON Um.id=m.UpdatedByUserId
	WHERE m.Id = ISNULL(@P_Id, m.Id) and  m.ViewYN = 1
		--AND m.ViewYN = 1
END
GO
```

# [dbo].[SP_GetMenuDetailsByUserId]

```sql
/****** Object:  StoredProcedure [dbo].[SP_GetMenuDetailsByUserId]    Script Date: 27-10-2023 15:08:31 ******/
SET ANSI_NULLS ON
GO
SET QUOTED_IDENTIFIER ON
GO
CREATE PROCEDURE [dbo].[SP_GetMenuDetailsByUserId] (@P_UserId INT)
AS
BEGIN
	SELECT RM.Id,
	RM.RoleId
		,R.[Name] [Role]
		,RM.MenuId
		,CASE WHEN m.Path LIKE'http%' THEN 'REPORTS' ELSE  PM.Name END ParentMenu
		,Case when m.MenuType='M' then 'Menu' when m.MenuType='S' then 'Sub Menu' else 'Item' end MenuType
		,M.[Name] [Menu]
		,m.Path
		,m.Icon
	FROM dbo.RoleMenu RM
	INNER JOIN Roles R ON RM.RoleId = R.Id
	INNER JOIN Menu M ON RM.MenuId = M.Id
	LEFT JOIN Menu PM ON M.ParentMenuId = PM.Id
	INNER JOIN Users u ON u.RoleId=r.Id
	WHERE u.Id = ISNULL(@P_UserId, u.Id)
	AND RM.ViewYN = 1
	AND ISNULL(m.Path,'')!=''
	AND m.Path not like 'http%'
	ORDER By r.name,CASE WHEN m.Path LIKE'http%' THEN 9999 ELSE  PM.[Order] END ,m.[Order]
END
GO
```
